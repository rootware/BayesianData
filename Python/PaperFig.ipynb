```python
import matplotlib.pyplot as plt;
import numpy as np;
```

```python
x = np.arange(-5.0,5.0,0.01);
y = np.cos( x-np.pi/2)**2
z = np.cos(x - np.pi/2+ 0.25*np.pi)**2
# z2 = 0.45* ( np.exp(- (x-np.pi)**2) + np.exp(- (x)**2)+ np.exp(- (x+np.pi)**2)  )
z2 = 0.75* (  np.exp(- (x)**2)  ) + 0.15 * np.cos(x)**2
```

```python
plt.figure(figsize =(14, 9));
plt.plot(x,y)
plt.plot(x,z,'--', color='r')
plt.fill_between(x, z2, color='lavender', alpha=0.9)
plt.axis('off')
```

```python

# Import libraries
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 
 
# Creating dataset
x = np.outer(np.linspace(-3, 3, 600), np.ones(600))
y = x.copy().T # transpose
# z = (np.sin(x **2/10.0) + np.cos(y **2) )
z = 4*np.exp(-(x **2+ y **2)/3) + 0.5*np.sin(x)+0.5*np.sin(y)
# Creating figure
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
 
# Creating plot
my_cmap = plt.get_cmap('coolwarm')

ax.plot_surface(x, y, z, cmap = my_cmap,edgecolor ='none', shade='red')
ax.view_init(50, 30) 
plt.grid(False)
plt.axis('off')
# show plot
plt.show()
```
